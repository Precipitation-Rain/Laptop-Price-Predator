```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('laptop_data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum().item()
```

```python
df.info()
```

```python
df['Ram'] = df['Ram'].str.replace('GB' , '')
df['Weight'] = df['Weight'].str.replace('kg' , '')
```

```python
df.head()
```

```python
df['Ram'] = df['Ram'].astype('int32')
df['Weight'] = df['Weight'].astype('float32')
```

```python
df.head()
```

```python
df.info()
```
